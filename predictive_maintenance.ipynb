# Enmarcar el problema. Panorama general
1.- Definir el objetivo en términos de negocio.
Se trata de predecir el fallo o no de una máquina en base a una serie de datos recogidos por diferentes sensores IoT.

2.- ¿Cómo se utilizará la solución?
La solución se usará para hacer mantenimiento preventivo. Las máquinas que obtengan fallo deberán ser revisadas de forma inmediata, mientras que aquellas que no lo tengan, podrán ser revisadas más adelante. De esta manera la planificación del mantenimiento es una tarea mucho más sencilla y eficaz que permite ahorrar costes de energía, prolonga la vida útil de los equipos y evita el mantenimiento innecesario.

3.- ¿Cuáles son las soluciones alternativas (si las hay)?
La alternativa es ajustar los intervalos de mantenimiento de los equipos en función de factores como la antigüedad, tiempo de uso e histórico de incidencias, entre otros.

4.- ¿Cómo debería plantear este problema (supervisado/ no supervisado, en línea/ fuera de línea, etc)?

El problema es **supervisado** porque los datos están **etiquetados**. Se va a hacer una **clasificación** en fallo/no fallo.

Las **características** van a ser:
* footfall: The number of people or objects passing by the machine.
* tempMode: The temperature mode or setting of the machine.
* AQ: Air quality index near the machine.
* USS: Ultrasonic sensor data, indicating proximity measurements.
* CS: Current sensor readings, indicating the electrical current usage of the machine.
* VOC: Volatile organic compounds level detected near the machine.
* RP: Rotational position or RPM (revolutions per minute) of the machine parts.
* IP: Input pressure to the machine.
* Temperature: The operating temperature of the machine.

Y el **objetivo** va a ser encontrar la variable fail (0/1)
fail: Binary indicator of machine failure (1 for failure, 0 for no failure).

5.- ¿Cómo medir el rendimiento?
A nivel de aplicativo el rendimiento lo vamos a medir por el número de aciertos según los casos de test.
A nivel de negocio para medir la mejora del rendimiento se pueden usar KPIs como
el MBTF que en el tiempo promedio entre fallos de las máquinas y se define como el tiempo total de operación/número de fallos. Se puede comparar este KPI entre mantenimiento tradicional y mantenimiento predictivo.

6.- Alineación con los objetivos empresariales
El objetivo empresarial es garantizar que la producción no se para debido a la avería de la maquinaria, sin que ello suponga un incremento elevado en los costes de mantenimiento. Habría que estudiar el coste actual del mantenimiento, teniendo en cuenta las pérdidas debidas a la avería de la maquinaria y el coste de la solución (inversión en sensores y software) junto con el coste de las pérdidas de esta solución de mantenimiento preventivo.

7.- ¿Cuál es el rendimiento mínimo para alcanzar el objetivo de negocio?
Suponemos que la inversión en hardware/ software inicial que se debe de hacer es de 100K. El coste/hora por parada debido a rotura de maquinaria es 10K. En el año 2024 tuvieron 20h de parada, es decir 200k. El año 2024 tuvo 12 máquinas paradas de 50, durante 20h. Negocio quiere disminuir como mínimo a la mitad las horas de parada. Esto supone que el nivel de acierto en la predicción se debe de situar en el 90%.

# Entrenamiento Machine Learning

```python
# importamos métricas
from sklearn.model_selection import cross_val_score, GridSearchCV
from sklearn.metrics import accuracy_score

# importar diferentes modelos de clasificación
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
```

```python
# simple training
knn_model = KNeighborsClassifier()

knn_model.fit(x_train, y_train)
```

```python
# predict data
y_pred = knn_model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

confusion_matrix(y_test, y_pred)
accuracy_score(y_test, y_pred)
```

```python
# entrenamos y evaluamos todos
# Crear listas para almacenar resultados
models = ['Logistic Regression', 'SVM', 'KNN', 'Decision Tree', 'Random Forest', 'Gradient Boosting']

accuracy_scores = []
precision_scores = []
recall_scores = []
f1_scores = []
roc_auc_scores = []

# Definir modelos
logistic_model = LogisticRegression()
svm_model = SVC(probability=True)  # Necesario para calcular AUC-ROC
knn_model = KNeighborsClassifier()
tree_model = DecisionTreeClassifier()
rf_model = RandomForestClassifier()
gb_model = GradientBoostingClassifier()

# Lista de modelos
model_list = [logistic_model, svm_model, knn_model, tree_model, rf_model, gb_model]

# Entrenar y evaluar cada modelo usando validación cruzada
for model in model_list:
    accuracy_scores.append(cross_val_score(model, x_train, y_train, scoring='accuracy', cv=5).mean())
    precision_scores.append(cross_val_score(model, x_train, y_train, scoring='precision', cv=5).mean())
    recall_scores.append(cross_val_score(model, x_train, y_train, scoring='recall', cv=5).mean())
    f1_scores.append(cross_val_score(model, x_train, y_train, scoring='f1', cv=5).mean())
    roc_auc_scores.append(cross_val_score(model, x_train, y_train, scoring='roc_auc', cv=5).mean())

# Crear DataFrame con los resultados
results_df = pd.DataFrame({
    'Model': models,
    'Accuracy': accuracy_scores,
    'Precision': precision_scores,
    'Recall': recall_scores,
    'F1 Score': f1_scores,
    'ROC AUC': roc_auc_scores
})

# Mostrar los resultados
print(results_df)
```
